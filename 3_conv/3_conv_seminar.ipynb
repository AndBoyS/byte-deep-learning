```python
class CustomConv2d:
    def __init__(
        self, 
        in_channels: int = 1, 
        out_channels: int = 1, 
        kernel_size: int = 3
        ) -> None:
        self.kernel_size = kernel_size
        self.weight = torch.ones((out_channels, in_channels, kernel_size, kernel_size))
        
    def __call__(self, tensor: torch.Tensor) -> torch.Tensor:
        batch_size, channel_size, feature_size_x, feature_size_y = tensor.shape[:2]
        
        num_iters_x = feature_size_x - self.kernel_size + 1
        num_iters_y = feature_size_y - self.kernel_size + 1
        
        output = torch.empty((
            batch_size, 
            channel_size, 
            num_iters_x,
            num_iters_y,
        ))
        
        
        
        
```
